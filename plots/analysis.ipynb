```python
import pandas as pd

# Load the birth_place_anchor_result.csv file
df = pd.read_csv('birth_place_anchor_result.csv')

# Count hallucination types
total_samples = len(df)
direct_hallucinations = (df['hallucination'] == 'direct').sum()
indirect_hallucinations = (df['hallucination'] == 'indirect').sum()
total_hallucinations = direct_hallucinations + indirect_hallucinations

# Calculate rates
direct_hallucination_rate = (direct_hallucinations / total_samples) * 100
indirect_hallucination_rate = (indirect_hallucinations / total_samples) * 100
total_hallucination_rate = (total_hallucinations / total_samples) * 100

# Display results
print("Birth Place Anchor Attack Results:")
print("=" * 50)
print(f"Total samples: {total_samples}")
print(f"\nDirect hallucinations: {direct_hallucinations}")
print(f"Direct hallucination rate: {direct_hallucination_rate:.2f}%")
print(f"\nIndirect hallucinations: {indirect_hallucinations}")
print(f"Indirect hallucination rate: {indirect_hallucination_rate:.2f}%")
print(f"\nTotal hallucinations: {total_hallucinations}")
print(f"Total hallucination rate: {total_hallucination_rate:.2f}%")
```
